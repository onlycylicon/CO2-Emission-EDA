```python
# import important packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# read and load dataset 
df_co2_emission = pd.read_csv('CO2 Emissions_Canada.csv')
```

```python
# display first 5 rows
df_co2_emission.head()
```

```python
#display last 5 rows
df_co2_emission.tail()
```

```python
# info about the dataset
df_co2_emission.info()
```

```python
# display rows and columns count
df_co2_emission.shape
```

```python
#quick statistics about the numerical columns
df_co2_emission.describe()
```

```python
# array of all total values in the column(Make)
df_co2_emission['Make'].unique()
```

```python
# counts of unique values in Engine Size
df_co2_emission['Engine Size(L)'].value_counts()
```

```python
# pivot table
df_co2_emission.pivot_table(index="Make",
                            values="Fuel Type",
                            aggfunc="sum")
```

```python
# Average Fuel Consumption in the city(L/100km)
df_co2_emission_car_type = df_co2_emission.groupby('Make')['Fuel Consumption City (L/100 km)'].mean()
print(df_co2_emission_car_type)

df_co2_emission_car_type.plot(kind='bar', xlabel='Car Brands',ylabel='Fuel Consumption',
                              title = 'Average Car Fuel Consumption City (L/100 km)')
#save plot
plt.savefig('acfcc.png')
#show plot
plt.show()
```

```python
## Average Fuel Consumption on the highway(L/100km)
df_co2_emission_car_type = df_co2_emission.groupby('Make')['Fuel Consumption Hwy (L/100 km)'].mean()
print(df_co2_emission_car_type)

df_co2_emission_car_type.plot(kind='bar', xlabel='Car Brand',ylabel='Fuel Consumption',
                              title = 'Average Car Fuel Consumption Hwy (L/100 km)')
#save plot
plt.savefig('acfch.png')
#show plot
plt.show()
```

```python
#Average Emission For Each Vehicle Class
df_co2_emission_car_type = df_co2_emission.groupby('Vehicle Class')['CO2 Emissions(g/km)'].mean()
print(df_co2_emission_car_type)

df_co2_emission_car_type.plot(kind='bar', xlabel='Vehicle Class',ylabel='CO2 Emissions(g/km)',
                              title = 'Average Emission For Each Vehicle Class')
#save plot
plt.savefig('aefevc.png')
#show plot
plt.show()
```

```python
#Total Emission By Vehicles
df_co2_emission_car_type = df_co2_emission.groupby('Make')['CO2 Emissions(g/km)'].sum()
print(df_co2_emission_car_type)

df_co2_emission_car_type.plot(kind='bar')

plt.title('Total Emission By Vehicle')
plt.xlabel('Car Brands')
plt.ylabel('Total')
#save plot
plt.savefig('total_emission.png')
#show plot
plt.show()

```
